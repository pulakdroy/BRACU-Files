```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
  # To Do
   self.head = Node(a[0],None,None)
   tail = self.head
   for i in range(1,len(a)):
     temp = Node(a[i],None,None)
     temp.prev = tail
     tail.next = temp
     tail = tail.next
   self.head.prev = tail
   tail.next = self.head
  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    # To Do
    count = 1
    k = self.head
    while k.next != self.head:
      count +=1 
      k=k.next
    return count
    
  
  # prints the elements in the list
  def forwardprint(self):
    # To Do
    k = self.head
    while k.next != self.head:
      print(k.element,end=",")
      k = k.next 
      
    print(k.element,end="")
    print()
    

  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    k = self.head.prev
    # temp = k
    while k != self.head:
      print(k.element,end=",")
      k = k.prev

    print(k.element,end="") 
    print()
    
      

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    count = 0
    k = self.head
    while k.next != self.head:
      if count == idx:
        return k
        break
      else:
        count += 1
        k = k.next
    return None
   
    
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    
    k = self.head.next
    count = 1

    while k != self.head:
      if k.element == elem:
        return count
      else:
        count += 1
        k = k.next
    return -1  
      
    

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
  #   # To Do
    k = self.head
    count = 0
    if idx<count or idx>self.countNode():
      return None
    elif idx == 0:
        new_node=Node(elem,k,k.prev)
        self.head.prev.next= new_node
        self.head.prev= new_node
        self.head= new_node
        return self.head
   
    else:
      for i in range (idx-1):
         k= k.next
      new_node= Node(elem, k.next,k)
      k.next.prev= new_node
      k.next = new_node
      return self.head
    

  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    # To Do
    k = self.head
    count = 0
    rev = None
    if idx<count or idx>self.countNode():
      return None
    elif idx == 0:
      rev= k.element
     
      k.next.prev= k.prev
      k.prev.next= k.next
      self.head= self.head.next
      return str(rev)
    else:
      for i in range(idx-1):
        k = k.next
      rev=k.next.element
      k.next= k.next.next
      k.next.prev= k
      return str(rev)
    
```
