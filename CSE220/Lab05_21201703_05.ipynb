```python
# Using Array based stack
class array:
    def __init__(self,e):
        self.array = [None]*e
        self.size = 0

    def push(self,elem):
        self.array[self.size] = elem
        self.size += 1

    def pop(self):
        new = self.array[self.size]
        self.array[self.size -1] = None
        self.size -= 1
        return new

    def peek(self):
        return self.array[self.size-1]

    def empty(self):
        empty = False
        if self.size < 1:
            empty = True
        return empty

v = input()
l = array(len(v))
count = 0
for i in v:
    flag = False
    if i == '[' or i == '{' or i == '(':
        l.push(i)
        if i == '[':
            num1 = count+1
        elif i == '{':
            num2 = count+1
        else:
            num3 = count+1
    elif i == ']' or i == '}' or i == ')':
        if l.empty() == True:
            flag = False
            break
        new = l.pop()
        if  (i == ']' and new == '[') or (i == '}' and new == '{') or (i == ')' and new == '('):
            flag = True
        else:
            i = new
            break
    else:
        flag = True
    count += 1
flag2 = False

if flag2 == True:
    if l.empty()==True:
        flag2 = True
    else:
        i = l.pop()
if flag == True and flag2 == True:
    print('This expression is correct')
elif flag == False or flag2 == False:
    print('This expression is not correct')
    if i == "]" or i == "}" or i == ")":
        print("Error at charcter #" + str(count) + ".", "'" + i + "'", "not opened.")
    elif i == "[" or i == "{" or i == "(":
        if i == "[":
            count = num1
        elif i == "{":
            count = num2
        else:
            count = num3
        print("Error at charcter #" + str(count) + ".", "'" + i + "'", "not closed.")
```

```python
#Using linked list based stack
class Node:
    def __init__(self,e):
        self.element = e
        self.next = None
class Link:
    def __init__(self):
        self.head = None
    def push(self,n):
        if self.head==None:
            self.head = Node(n)
        else:
            new = Node(n)
            new.next = self.head
            self.head = new
    def peek(self):
        return self.head.element
    def pop(self):
        if self.head == None:
            return None
        else:
            popped = self.head.element
            self.head = self.head.next
            return popped
    def empty(self):
        empty = False
        if self.head == None:
            empty = True
        return empty    

v = input()    
v1 = Link()
count = 1
for i in v:   
    flag = False
    if i == "[" or i == "{" or i == "(":
        v1.push(i)
        if i == "[":
            num1 = count
        elif i == "{":
            num2 = count
        else:
            num3 = count
    elif i == "]" or i == "}" or i == ")":
        
        if v1.empty() == True:
            flag = False
            break
        temp = v1.pop()
        if (i == "]" and temp=="[") or (i == "}" and temp=="{") or (i == ")" and temp=="("):
            flag = True
        else:
            i = temp
            break
    else:
        
        flag = True    
    count+=1
flag2 = False  
if flag == True:
    if v1.empty()==True:
        flag2 = True
    else:
        i = v1.pop()    
if flag == True and flag2 == True:
    print("This expression is correct")
elif flag == False or flag2 == False:
    print("This expression is NOT correct")
    if i == "]" or i == "}" or i == ")":
        print("Error at charcter #"+str(count)+".","'"+i+"'","not opened.")
    elif i =="[" or i =="{" or i =="(":
        if i  == "[":
            count = num1
        elif i == "{":
            count = num2
        else:
            count = num3
        print("Error at charcter #"+str(count)+".","'"+i+"'","not closed.")
        
```
