```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):
  # TO DO
  # Hint, You can write a function for left shift once and then use it
   for j in range(k):
        for i in range(len(source)-1):
            source[i] = source[i+1]
        source[len(source)-1] = 0
   return source


# Test 02: Rotate Left k cell
def rotate_left(source, k):
  # TO DO
  # Hint, You can write a function for left rotate once and then use it
  
  for i in range(k):
    temp = source[0]
    for j in range(len(source)-1):
      source[j] = source[j+1]
    source[len(source)-1] = temp    
  return source


# Test 03: Shift Right k cell
def shift_right(source, k):
  # TO DO
  # Hint, You can write a function for right shift once and then use it
  # for i in range(k):
  for i in range(k):
        temp = len(source)-2
        for j in range(temp,-1,-1):
            source[j+1] = source[j]           
        source[0] = 0
  return source



# Test 04: Rotate Right k cell
def rotate_right(source, k):
  # TO DO
  # Hint, You can write a function for right rotate once and then use it
  for i in range(k):
        temp = source[len(source)-1]
        for j in range(len(source)-2,-1,-1):
            source[j+1] = source[j]
        source[0] = temp
  return source



# Test 05: Remove an element from an array
def remove(source, idx):
  # TO DO
  for i in range(idx+1,len(source),1):
    source[i-1] = source[i]
  return source



# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source,element):
  i=0
  while i<len(source):

    if source[i] == element:
      
      for j in range(i,len(source)-1):
            source[j] = source[j+1]
      source[len(source)-1] = 0
      i-=1
    i+=1
  return source   
  



# Test 07: Splitting an Array
def split_array(source):
  # TO DO
  sum = 0
  flag = False
  for i in range(len(source)):
    sum+=source[i]
    sum2 = 0
    for j in range(i+1,len(source)):
      sum2+=source[j]
    if sum2 == sum:
      flag = True
  return flag


# Test 08: Max Bunch Count
def max_bunch(source):
  #TO DO
  counter = 1
  max = 0
  for i in range(len(source)-1):
    if source[i] == source[i+1]:
      counter+=1
      if counter>max:
        max=counter
    else:
        counter = 1
  return max

```

```python
#PART 2
from math import sqrt 
def array(source):
    sum = 0
    for i in source:
        sum += i       

    mean = sum / len(source)
    return mean


def standard(source):
     sum = 0
     sum1 = 0
     for i in source:
        sum += i       

     mean = sum / len(source)
     for j in source:
       sum1 += (j-mean)**2
     sd = sqrt((sum1)/(len(source)-1))
     return sd

def new_array(source):
     sum = 0
     sum1 = 0
     new_arr = []
     for i in source:
        sum += i       

     mean = sum / len(source)
     for j in source:
       sum1 += (j-mean)**2
     sd = sqrt((sum1)/(len(source)-1))
    
     for k in source:
         if  k > mean + sd * 1.5 or k < mean - sd*1.5: 
           new_arr.append(k)
     
     return new_arr  

source = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]
print(array(source))
print(standard(source))
print(new_array(source))
```
